# Лемматизируем текст

```python
# для очистки (препроцессинга) текста...
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
ru_stop_words = stopwords.words('russian') # давайте для русского сохранять в отдельную переменную
stop_words = stopwords.words('english')
import spacy

# pymorphy
!pip install pymorphy2
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()

!pip install spacy
!python -m spacy download en_core_web_sm
```

```python
# если у вас русский, запустите только команду для русского
# английский пропустите

def get_ru_lemmas(text):
  text = text.lower()
  text_tokens = word_tokenize(text)
  clean_tokens = []
  for i in text_tokens:
    if i[0].isalpha() and i not in ru_stop_words:
      clean_tokens.append(i)
  ru_lemmatized = []
  for word in clean_tokens:
    result = morph.parse(word)
    most_probable_result = result[0]
    normal_form = most_probable_result.normal_form
    ru_lemmatized.append(normal_form)
  return ru_lemmatized
```

```python
nlp = spacy.load("en_core_web_sm")

def get_eng_lemmas(text):
  text = text.lower()
  text_tokens = word_tokenize(text)
  clean_tokens = [] # чистим список токенов
  for i in text_tokens:
    if i[0].isalpha() and i not in stop_words:
      clean_tokens.append(i)
  spacy_token = []
  doc = nlp(' '.join(clean_tokens))
  for token in doc:
    if token.lemma_ != '-' and token.lemma_ != '.':
        spacy_token.append(token.lemma_)
  return spacy_token
```

```python
def parser(dict_person, text):
  persons = []
  for word in get_ru_lemmas(text): # можно заменить на get_eng_lemmas
    for key in dict_person:
      if word in dict_person[key]:
        persons.append(key)
  return list(set(persons))

#[[Персонаж1, Персонаж2, Персонаж3], [], [], [Персонаж1]] - в таком виде они вернутся, это нормально
```

```python
result = []
for i in text_list:
  if parser(dict_person_clean, i) != None:
    result.append(parser(dict_person_clean, i))

result
```

```python
def network_creator(result):
  p = []
  pair = []
  for item in result:
      for i in range(len(item)):
        for j in range(i + 1, len(item)):
          pair.append([item[i], item[j]])
  for i in pair:
    i.sort()
  pair_tuple = []
  for i in pair:
    if tuple(i) not in pair_tuple:
      pair_tuple.append(tuple(i))
  subsets = {t : set(t) for t in pair_tuple}
  mainsets = [set(xs) for xs in pair]
  pairs = {
    tup : sum(s.issubset(m) for m in mainsets)
    for tup, s in subsets.items()
  }
  sorted_dict = {}
  sorted_keys = sorted(pairs, key=pairs.get, reverse=True)
  for w in sorted_keys:
      sorted_dict[w] = pairs[w]
  return sorted_dict

network = network_creator(result)
network

# [[Персонаж1, Персонаж2, 1], [Персонаж1, Персонаж3, 1], [Персонаж2, Персонаж3, 1]] - подсчитаны взаимодействия
```

```python
graph_l = []
g = []
for i in network_creator(result):
  g.append(i[0])
  g.append(i[1])
  g.append(network_creator(result)[i])
  graph_l.append(tuple(g))
  g = []
set(graph_l)
print(graph_l)

graph_list = []
for i in graph_l:
  a = [i[0], 'undirected', i[1], i[2]]
  graph_list.append(a)
print(graph_list)
```
