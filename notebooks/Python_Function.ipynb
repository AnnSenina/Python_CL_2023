Функция в python - это такой объект, который выполняет какое-то действие c данными, которые она принимает на вход.

* То, что функция принимает, называется аргументом и записывается в скобках. У функции может быть как один, ноль или несколько аргументов.

* То, что функция возвращает, называется значением.

Обычно функция определяется с помощью инструкции def, за которым следует название функции со скобками и двоеточием.



Кстати, мы с самого начала работали с функциями - стандартными, которые Python уже знает:

print(), input(), len(), max(), min(), sum()

```python
# Давайте создадим (определим) самую простую функцию:

def hello(): # нет аргументов
    print("hello \n(∩｀-´)⊃━☆ﾟ.*･｡ﾟ")

hello()
```

```python
#  усложним функцию

def personal_hello():
    name = input("Как Вас зовут?")
    print("Добрый день,", name)
    print()
    # ничего не возвращает

personal_hello()
```

Можно написать функцию так, чтобы она не только совершала действие, но и возвращала некоторый результат. Понадобится ключевое слово return



```python
# функция с возвратом значения

def add(x, y): # принимает аргументы
    res = x + y
    return res

print(add("h", "i")) # так мы вызвали функцию и передали аргументы в скобках
print(add(7, 10))
print(add(['Anna', 'Iliana'], ['Jane', 'Jack']))
l = add(['Anna', 'Iliana'], ['Jane', 'Jack']) # сохраним в переменную
print(l)
```

```python
def unique(x): # простая функция, выдающая уникальный набор элементов в объекте
    x = set(x)
    return x

print(unique("hello"))

poem  = "The Rose is a rose and was always a rose."
print(unique(poem))

poem = poem.strip(".-:!?–,").lower().split() #предобратобка (как это работает, посмотрим отделельно)
print(unique(poem)) # множество уникальных токенов
```

##бонус: lambda func
Анонимные функции:

могут содержать лишь одно действие
- выполняются быстрее
- не обязательно присваивать переменной (не нужен def)
- не требуется return
- обычно используются для одноразовых функций

создаются с помощью ключевого слова lambda

```python
lambda x, y: x + y  # задали (до двоеточия аргументы, после - действие с ними)
func = lambda x, y: x + y # так можно сохранить в переменную
print(func(4, 7)) # вызываем, передав аргументы

#def название_функции(x, y):
#  return x + y

# аналогично lambda x, y: x + y
```

```python
func = lambda x, y: x + y
print(func("i write"," code in Python")) # подаем строки как аргументы, их тоже можно конкатенировать
```

##Важно!
lambda-функции отлично работают в сочетании с сортировками

Как сортировать разные объекты?

- функция sorted() - не изменяет исходный объект
- метод .sort()

```python
# по умолсанию числа сортируются в порядке возрастания
a = [1, 7, -500, 956, 10000, 52]
print(sorted(a))
print(a)
```

```python
# по умолчанию строки сортируются по алфавиту
a = ['hello', 'python', 'Programming', 'functions', 'lambda']
print(sorted(a))
```

```python
# можно развернуть: отсортируй в обратном порядке, используя аргумент reverse
a = [1, 7, -500, 956, 10000, 52]
print(sorted(a, reverse=True))

b = ['hello', 'python', 'Programming', 'functions', 'lambda']
print(sorted(b, reverse=True))
```

```python
# как отсортировать без учета регистра?
a = ['hello', 'python', 'Programming', 'functions', 'lambda']
print(sorted(a, key = lambda x: x.lower()))
```

Функцией sorted можно сортировать списки, кортежи, множества, словари (по ключам)

Функция sorted всегда возвращает отсортированный **список**

Зачем тогда метод .sort?

Метод изменяет исходный объект, это метод списков (с кортежами не работает)

```python
a = [1, 7, -500, 956, 10000, 52]
a.sort(reverse=True)
print(a)
```

Иногда нужно сортировать список списков, причем не по нулевому индексу - на помощь приходит параметр сортировки key, в который можно передать lambda-функцию

```python
names = [['Ann', 33],
         ['Kate', 50],
         ['Mike', 16],
         ['Dan', 44]]
print(sorted(names)) # по умолчанию = по нулевому индексу => по алфавиту

# сделаем по первому индексу - по возрасту

print(sorted(names, key=lambda x: x[1])) # правило сортировки - по 1 индексу
```
