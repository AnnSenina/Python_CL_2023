```python
import random
import turtle

# Инициализация экрана
screen = turtle.Screen()
screen.bgcolor("black")

# Создание черепашки
t = turtle.Turtle()
t.speed(0)
t.width(2)


def draw_galaxy(x, y, size):
    colors = ["red", "orange", "yellow", "green", "blue", "purple"]

    t.up()
    t.goto(x, y)
    t.down()

    for i in range(size):
        t.color(random.choice(colors))
        t.forward(i)
        t.left(59)


def draw_starry_sky():
    colors = ["white", "yellow", "cyan", "magenta"]

    for _ in range(300):
        x = random.randint(-400, 400)
        y = random.randint(-400, 400)
        size = random.randint(1, 5)

        t.up()
        t.goto(x, y)
        t.down()

        t.color(random.choice(colors))
        t.begin_fill()
        t.circle(size)
        t.end_fill()


def draw_planet(x, y, radius):
    colors = ["#FF6347", "#FFA500", "#FFFF00", "#00FF00", "#00FFFF", "#8A2BE2", "#FF00FF", "#FF69B4"]

    t.up()
    t.goto(x, y - radius)
    t.down()

    t.color(random.choice(colors))
    t.begin_fill()
    t.circle(radius)
    t.end_fill()


def generate_art():
    t.reset()
    t.hideturtle()
    t.speed(0)


    draw_starry_sky()


    for _ in range(8):
        x = random.randint(-screen.window_width() // 2, screen.window_width() // 2)
        y = random.randint(-screen.window_height() // 2, screen.window_height() // 2)
        radius = random.randint(10, min(screen.window_width(), screen.window_height()) // 16)

        draw_planet(x, y, radius)

        x1 = random.randint(-screen.window_width() // 2, screen.window_width() // 2)
        y1 = random.randint(-screen.window_height() // 2, screen.window_height() // 2)
        radius1 = random.randint(10, min(screen.window_width(), screen.window_height()) // 16)

        draw_galaxy(x1, y1, radius1)


    turtle.done()



generate_art()
```
