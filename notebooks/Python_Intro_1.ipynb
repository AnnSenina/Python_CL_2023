###Переменные

Результат подсчета (и любые другие данные) можно сохранять в переменные

Переменные хранят данные, мы присваиваем им значение через оператор =

Не путайте операторы = и ==

```python
n = 10
# мы создали переменную n и присвоили ей значение 10
print(n)
# Python ее запомнил

print(n == 10.0) # сравнение 10 и 10.0
print(n == "10") # сравнение 10 и "10"

# важное о вещественных числах (дробях)
a = 0.1
b = 0.2
print(a + b) # результат - не красивая десятичная дробь, но что-то очень близкое
c = 0.3
print(a + b == c) # ложь
```

```python
print(a + b + c)
```

```python
k = 0
print(k + n) # ошибка имени
```

###Операции со строками

1. строки можно конкатенировать, присоединять друг к другу

```python
line1 = "я позвал в кино " # создали первую переменную
line2 = "этих людей " # создали вторую

concat = line2 + line1 # новая переменная, в нее сохраним результат конкатенации первых двух переменных
another_concat = line1 + line2 # а это еще одна, мы снова конкатенируем, но в обратном порядке (!)

# результаты!
print(concat)
print(another_concat)
```

4. доступ по индексу

строка - это итерируемый объект (= состоит из элементов, как слово из букв. Мы можем пройтись по каждому из его элементов). Мы можем работать с элементами строки. Для этого нам понадобятся индексы, порядковые номерá элемента. Индекс записывается в квадратных скобках [ ]

!! нумерация элементов начинается с 0, т.е. первый элемент имеет индекс [0].

Легкое правило запоминания - индекс всегда на 1 меньше чем настоящий номер элемента

```python
new = "мы пришли, чтобы кодить"
print(new[0]) # хотим напечатать первый элемент в сторке
print(new[3]) # печатаем 4й элемент
print(new[-1])# хотим последний элемент, он же первый с конца
```

4.1. Извлечение среза


Иногда вам может понадобиться не вся последовательность элементов, а какой-то ее кусочек. Такой кусочек называниется срез (англ. slice)

Оператор извлечения среза: [x:y], где x – начало среза, а y – окончание;

Обратите внимание, что **символ с номером y (верхняя граница) в срез не входит**.

По умолчанию первый индекс равен 0, а второй - длине строки.

```python
print(new) # сначала напечатаем строку целиком

print(new[3:9]) # а теперь напечатаем элементы строки, с четвертого по девятый
```

```python
# можем сохранить слайс/срез в новую переменную
piece = new[3:-1] # сохраним элементы с четвертого до последнего
print(piece) # давайте напечатаем сохраненное на предыдущем шаге

print(new[:3]) # напечатаем слайс/срез элементов с первого до третьего, третий не напечатается

print(new[3:]) # напечатаем слайс/срез элементов с четвертого и до конца

print(new[:]) # напечатаем слайс/срез элементов, но так как они не заданы, они берутся по дефолту - первый и последний
```

```python
s = 'hello' # неизменяемый тип данных
s[1] = 'a'
print(s)
```

```python
s = s[0] + 'a' + s[2:]
print(s)
```

4.2. можно задать шаг, с которым нужно извлекать срез/слайс. Шаг обозначается третьим в квадратных скобках

[начало:конец:шаг]

```python
# напечатаем строку с первого до последнего элемента(т.к. индексы не заданы), но шаг в обратном порядке
print(new[::-1])

# с первого до последнего элемента(т.к. индексы не заданы), шаг в прямом порядке, через три
print(new[::3])

# напечатаем строку с третьего до последнего элемента(т.к. индексы не заданы), но шаг в обратном порядке
print(new[2::-1])
```
