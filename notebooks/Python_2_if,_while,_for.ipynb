# Условия (if-elif-else)

if (оператор ветвления)

Проверяет условие на правдивость.

Общая схема:

- if с проверкой условия
- необязательно: elif (иначе-если), доп.условия
- в конце необязательная часть else (во всех остальных случаях)



Условия.png

```python
# пример:
inp = input("введите целое число, а я угадаю, четное оно или нет:")

inp = int(inp) # преобразуем строку в число

# Дальше создадим условие:

if inp % 2 == 0:
    print('да, число', inp, "четное")
else:
    print('число', inp, 'нечетное')
```

```python
age = int(input("Введите возраст ребенка")) # детские кружки
if age <= 6:
  print("Вам доступны музыка, рисование")
elif age <= 10: # каждый if независим! не начинаем новое условие, а продолжаем конструкцию - используем elif
  print("Вам доступны спорт, шахматы")
elif age <= 14:
  print("Вам доступны программирование, робототехника")
else:
  print('Для детей старше 14 лет у нас пока нет кружков')
```

```python
# диалог о погоде

a = input("Какая сейчас погода?") # солнце, дождь, снег

if a == "солнце":
  print("Возьмем солнечные очки")
elif a == "дождь":
  print("Возьмем зонт")
elif a == "снег":
  print("Наденем шарф")
else:
  print("Одевайтесь по погоде")
```

# Цикл while

Цикл whilr - это многократный if (пока условие выполняется.. следуем инструкциям цикла; когда условие не выполнено, цикл заканчивается)

цикл.png

```python
n = 10
while n > 0:
  print(n) # 10, 9... 1

  n = n - 1 # правило для остановки
```

```python
n = 10
while True: # всегда
  print(n)
  n = n - 1

  if n == 0:
    break # ломаем цикл
```

```python
i = 1 # вне цикла создали переменную, равную 1

while i < 10: # начало цикла: до тех пор, пока i меньше 10

    print(i) # тело цикла: печатаем i
    i = i + 2 # тело цикла: пересохраняем значение i, прибавив к нему 2, повторяем сначала

print("Done!")
```

### range(от, до, шаг)

У каждого списка есть индексы - можно попросить пайтон напечатать все индексы списка:

```python
for i in range(len(c)): # range - генерирует последовательность чисел
  # len - считатет длину
  print(i) # индексы
```

```python
# объединим
for i in range(len(c)): # i = 0, 1, 2, 3
  print(c[i], 'печатаем элемент')
  print(i, 'печатаем его индекс')
```

Если мы хотим внести изменения в исходный список, нам недостаточно доставать оттуда значения:

мы должны обратиться к элементу по индексу и перезаписать его

```python
print(c)

for i in range(len(c)): # i = 0, 1, 2, 3
  c[i] = c[i] + 100

print(c)
```

```python
# а range откуда пришел? он не страшный :)
# генерирует числовые последовательности

print(*range(10)) # от 0 по умолчанию до 9

print(*range(1, 10)) # указать: от 1 до 9

d = list(range(10, 50)) # от 10 до 49 и преврати в список
print(d)

print(list(range(0, 100, 5))) # от 0 до 99 с шагом в 5
# даже умеет делать это с шагом, как в индексации строк и списков

# range(len(список)) == "сгенерируй последовательность от 0 до длины списка" == индексы
```

### Бонус: генераторы текстов

Давайте создадим генератор фраз с помощью работы со списками, циклами и библиотеки random



```python
import random
print(random.randint(0, 100)) # случайное число
```

```python
my_list = ['один', 'два', 'три']
print(random.choice(my_list))
```

```python
import random # чтобы работать с библиотекой, ее надо импортировать

# генератор одной фразы

attributes = ['boring','sad','risky','silly','difficult','easy']

drinks = ['coffee','tea','wine','juice',"milk","water"]

phrase = "They never drink "+ random.choice(drinks) + " before 9 am because it's too "


reason = random.choice(attributes)

result = phrase + reason
print(result)
```

```python
import random # чтобы работать с библиотекой, ее надо импортировать

# генератор фраз
attributes = ['boring','sad','risky','silly','difficult','easy']

drinks = ['coffee','tea','wine','juice',"milk","water"]

for i in drinks: # для каждого элемента (i) в списке drinks:

    phrase = "They never drink "+ i +" before 9 am because it's too "
    # сохраним в переменную phrase такую cтроку

    reason = random.choice(attributes) # сохраним в переменную reason рандомный выбор одного из элементов списка seq2

    result = phrase + reason
    print(result)
```
