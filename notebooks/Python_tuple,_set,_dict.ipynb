# Кортежи

## Кортеж (tuple)
это неизменяемый список.

Зачем нужны кортежи, если есть списки?

1. Защита данных. Кортеж защищен от изменений, как намеренных (что плохо), так и случайных (что хорошо).
2. Меньший размер. Вам нужно меньше памяти, чтобы хранить данные

```python
# сравним размеры в байтах

a = (1, 2, 3, 4, 5, 6) # кортеж
b = [1, 2, 3, 4, 5, 6] # лист/список

print("Size of the tuple: ",a.__sizeof__(),"\nsize of the list: ",b.__sizeof__())
```

```python
# как создать кортеж
a = tuple()
print(a, "\n",type(a))

# с помощью литерала кортежа ()
b = ()
print(b, "\n",type(b))
```

```python
# Создаем кортеж из одного элемента:
a = ('apple', )
print(a)
b = 'apple', # в этом случае скобки разрешается не ставить
print(b)

c = ('apple') # строка
print(c)
```

кортежам доступны все операции над списками, не изменяющие список (об этом поговорим на следующей паре)



```python
# проверим, что кортежи не изменяются:
b = ('apple', )
b[0] = 'banana'
print(b)
# ошибка - кортеж нельзя изменить
```

```python
b = ['apple', 'lemon']
b[0] = 'banana'
print(b)
```

## Множество (set)
это тоже "контейнер" для данных, но в отличие от листа(списка), он содержит уникальные, не повторяющиеся элементы.


```python
# создать множество можно только с помощью функции set()
my_set = set()
print(my_set)

# а вот так нельзя:
# my_set = {} # (со списками, кортежами и словарями этот способ сработает)
# скобки - одинаковые со словарями, питон по пустому содержимому не понимает и создает словарь
```

```python
print(set([34, 78, "banana"])) # объекты перемешались, индексы не работают

print(set("hello there")) # строка - перебираемый объект, распалась на буквы
# повторы исчезли, буквы перемешались

tmp = ['hello', 'hello', 'there']
print(set(tmp))
```

```python
# множества удобно использовать для удаления повторяющихся элементов:

fruits = ["apples", "peaches", "cherries", "pears", "oranges", "persimmons", "peaches"]
# задали список, сохранили в переменную

# давайте изменим тип объекта на set с помощью функции set()
fruits = set(fruits) # применили функцию, пересохранили переменную

print(fruits) # персики больше не дублируются
```

```python
cars = ["BMW", "Toyota", "Bentley", "Jeep", "Jeep"]

cars = set(cars)

#число (уникальных) элементов в множестве (размер множества).
print(len(cars))
```

```python
# принадлежит ли элемент множеству (со списками, кортежами тоже работает)

print("BMW" in cars)
print("Chevrolet" in cars)
```

### Логические операции

```python
fruits = {'apples', 'cherries', 'oranges', 'peaches', 'pears', 'persimmons'}
fruits2 = {'apples', 'cherries', 'oranges', 'peaches', 'pears', 'persimmons', 'apples'}

berries = {"strawberry", "red currant"}

# сравнение множеств
print(fruits == berries)
 # True если все элементы одного принадлежат второму
print(fruits == fruits2)
```

```python
cars = {"BMW", "Toyota", "Bentley", "Jeep", "Jeep"}
french_cars = {"Peugeot", "Citroën", "Renault", "Jeep"}

print(cars & french_cars) # найди общее
print(cars | french_cars) # найди все возможные варианты
print(cars ^ french_cars) # только различное, симметрическая разница
print(cars - french_cars) # найди то, что есть в первом множестве, но чего нет во втором
```

# Словарь (dict)
коллекции произвольных объектов, записанные в виде пары ключ : значение (ключи уникальные, значения могут повторяться)

(словари = ассоциативные массивы = хеш-таблицы)

Давайте создадим словарь:

```python
#  помощью литерала {}
d = {}
print(d, type(d))
```

```python
d_names = {"Jane": ["London", 35],
           'Olga': ["Moscow", 40],
           "Mike": ["Spb", 26]
           }
# Jane, Olga... - это ключи, списки - значения
print(d_names)
```

```python
# с помощью функции dict()
d = dict(A = 67, B = 903)
print(d)
```

```python
a = ["Jane", "Oleg", "Mia"]
b = [25, 27, 56] # только списки одного размера

res = zip(a, b) # функция, которая создает словарь из двух списков

print(dict(res))
```

```python
# получить значение по ключу (аналогично индексам)
d = {1: "A", 2: "B", 3: "C"}
print(d[2]) # достаем значение по ключу 2 (не индекс! индексов нет)
```

```python
d = {1: "A", 2: "B", 3: "C"}

d[4] = "D" # добавляем  новую пару, где ключ = 4, а значение = D
print(d) # посмотрим на весь словарь
```

```python
# аналогично добавим новую пару в словарь с именами
d_names = {"Jane": ["London", 35], 'Olga': ["Moscow", 40], "Mike": ["Spb", 26]}

d_names["Jack"] = ["Toronto", 35]
print(d_names["Jack"])
print(d_names)
```

```python
# перезапишем значение по ключу
d = {1: "A", 2: "B", 3: "C"}

print(d[2])
d[2] = "W" # если такой ключ уже есть, он перезапишется
print(d[2])
print(d)
```

```python
# присвоение по новому ключу расширяет словарь;
# присвоение по существующему ключу перезаписывает его значение;
# попытка извлечения несуществующего ключа порождает исключение (KeyError).
```

### Все контейнеры можно перебирать циклами

```python
my_tuple = (12, 556, 3348)
for i in my_tuple:
  print(i)
  print(i + 1000) # можно что-то менять перед печатью
```

```python
my_set = {'apple', 'banana', 'pear'}
for i in my_set:
  print(i) # перебираются, но в беспорядке
```

```python
d = {'Jane': "London", 'Olga': "Moscow", "Paul" : "Paris", "Ivan" : "Minsk", "Anne" : "Copenhagen"}

for i in d:
  print(i, 'ключ')
  print(d[i], 'значение')
```
